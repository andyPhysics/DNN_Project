```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.loadtxt('file1.txt')
y = np.loadtxt('file2.txt')
z = np.loadtxt('file3.txt')
```

```python
epoch1 = x[0]
loss1 = x[1]
val_loss1 = x[2]
epoch2 = y[0]
loss2 = y[1]
val_loss2 = y[2]
epoch3 = z[0]
loss3 = z[1]
val_loss3 = z[2]
```

```python
plt.figure(1)
plt.plot(epoch1,loss1,'.') #This is normalized between 0 and 1
plt.plot(epoch2,loss2,'x') #This is normalized between -0.5 and 0.5
plt.plot(epoch1,val_loss1,'o')
plt.plot(epoch2,val_loss2,'--')
plt.plot(epoch3,loss3)
plt.plot(epoch3,val_loss3)
```
